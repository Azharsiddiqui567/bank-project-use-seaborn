```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
a = sns.load_dataset("titanic")
a
```

```python
sns.barplot(x='who', y='age', data=a)
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot')
```

```python
sns.barplot(x='alone', y='fare', data = a)
```

```python
sns.barplot(x='alone', y='fare', data = a, hue='embark_town')
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot', order=['child','man', 'woman who'])
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot', estimator=np.min)
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot', estimator=np.max)
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot', estimator=np.mean)
```

```python
sns.barplot(x='who', y='age', data=a, hue='sex', palette='hot', estimator=np.median)
```
